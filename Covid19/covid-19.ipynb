```python
create = pd.io.sql.get_schema(df.reset_index(), 'covid19', con=source)
with open('covid19.sql', 'w+') as file:
    file.write(create)
    file.close()
    
def populate():
    DDls = 'INSERT INTO test.covid19 (`'+ str('`, `'.join(df.columns))+ '`)\nVALUES '
    for index, row in df.iterrows():   
        if index == len(df) - 1:
            DDls += str(tuple(row.values)) + ';' 
        else:
            DDls += str(tuple(row.values)) + ',\n'
    return DDls

with open('covid19.sql', 'a+') as file:
    file.write(populate())
    file.close()
```
